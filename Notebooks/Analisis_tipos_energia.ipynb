```python
def calidaDato (data):
    df = data
    calidad = []
    mascara = df.isna().sum()
    for i in range(0,len(mascara)):
        p = (df.shape[0]- mascara[i]) / df.shape[0]
        calidad.append([mascara.index[i], round((p*100),2)])
   
    calidad = pd.DataFrame(calidad)
    calidad.rename(columns={0:'Columna', 1: 'Calida %'}, inplace=True)
    
    return calidad
```

```python
calidad = calidaDato(df)
```

```python
def mapPlot(df_newmap):
    data = dict(type = 'choropleth',
                locations = df_newmap['País'], 
                locationmode = 'country names', 
                z = list(df_newmap['Renovables (% Energía Primaria Equivalente)']), 
                colorscale='tropic',
                text = df_newmap['País'])
    layout = dict(geo = dict(scope='world'))

    fig = go.Figure(data=[data], layout=layout)
    fig.update_layout(
        margin=dict(l=5,r=5,b=5,t=5),
        paper_bgcolor= 'black',
        font_color='#cee3e1',
    )              
    return fig
```

```python
def linealPlot(df_new):
    fig= go.Figure(data=[
        go.Bar(
        x=df_new.País,
        y=df_new["Renovables (% Energía Primaria Equivalente)"],
        name='dif 2'
        )
    ])
    fig.update_layout(
        height=300, width=1000,
        plot_bgcolor='black',
        paper_bgcolor= 'black',
        font_color='#cee3e1',
        legend=dict(
            x=0.05,
            y=1,
            title_font_family="Times New Roman",
            font=dict(
                family="Courier",
                size=12,
color="LightSteelBlue"
            ),
            bgcolor="Black",
            bordercolor="LightSteelBlue",
            borderwidth=1
        ),
        xaxis=dict(showgrid=False,showline=True,linecolor='rgb(255,255,255)'),
        yaxis=dict(showgrid=False),
        margin=dict(l=10,r=10,b=10,t=10)
    )
    return fig
```
