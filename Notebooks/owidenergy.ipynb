```python
#validación de datos no nulos
diccionario = {}
mascara = df_owid.isna().sum()
for i in range(0,len(mascara)):
    p = (df_owid.shape[0] -mascara[i])/df_owid.shape[0]
    if (p <0.5) & (p>0.39):
        diccionario[mascara.index[i]] = round(p,2)
    
diccionario
```

```python
#combustibles fósiles
df_owid.loc[:,['fossil_electricity','oil_electricity','coal_electricity','gas_electricity']].sum()
```

Las columnas de energías fósiles:
- fossil_electricity = oil_electricity  + coal_electricity + gas_electricity

```python
#combustibles renovables
renovable = df_owid.loc[:,['renewables_electricity','wind_electricity','solar_electricity',
    'hydro_electricity','biofuel_electricity',
    'other_renewable_electricity','other_renewable_exc_biofuel_electricity','low_carbon_electricity']].dropna()

renovable[renovable.renewables_electricity >30]
```

Las columnas de energías renovables:
- renewables_electricity = wind_electricity + solar_electricity + hydro_electricity + other_renewable_electricity
